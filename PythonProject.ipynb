```python
# Import the pandas library, which is essential for working with dataframes
# in Python.

import pandas as pd

# Read the 'customer_shopping_behavior.csv' file into a pandas DataFrame.
# A DataFrame is like a table with rows and columns, similar to a spreadsheet.
# This line loads all the data from the CSV file into a variable named 'df'
# (short for DataFrame).

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
# This line shows the first 5 rows of the DataFrame, 'df'.
# It's a quick way to look at the data structure and content after loading it.
df.head()
```

```python
# This line displays the first 5 rows of the DataFrame, 'df'.
# It helps you quickly see the data's structure and contents.



df.info()
```

```python
# This line generates descriptive statistics for all columns in the DataFrame 'df'.
# For numerical columns, it provides count, mean, standard deviation, min,
# max, and quartiles.
# For categorical (object) columns, it shows count, unique values, top
# occurring value, and its frequency.
# 'include="all"' ensures that both numerical and object type columns
# are included in the summary.


df.describe(include="all")
```

```python
# This line checks for missing values (NaN or null) in each column of the DataFrame 'df'.
# `df.isnull()` just makes a new table that tells you exactly where all the empty
# spots (True) are in your original data.
# `.sum()` then counts the number of True values (missing entries) for each column.
# This helps in quickly identifying columns with incomplete data.



df.isnull().sum()
```

```python
# This line fills in any missing 'review_rating' values.
# It first groups the data by 'category' (like Clothing, Footwear, etc.).
# Then, for each category, it calculates the median (middle) 'review_rating'.
# Finally, it fills any missing ratings in that category with that category's median rating.
# This ensures that missing values are replaced with a reasonable value based on similar items.





df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# This line converts all column names in the DataFrame to lowercase.
df.columns = df.columns.str.lower()



# This line replaces any spaces in the column names with underscores to make them easier to work with.
df.columns = df.columns.str.replace(" ","_")



# This line renames a specific column 'purchase_amount_(usd)' to 'purchase_amount' for simplicity.
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# This line displays all the current column names of the DataFrame 'df'.
# It's useful to verify that the column renaming and formatting (lowercase, underscores) worked
# as expected.




df.columns
```

```python
# This line creates a Python list called `labels`.
# This list contains descriptive names for different age categories:
# 'Young Adult', 'Adult', 'Middle-Age', and 'Senior'.
# These labels will likely be used later to categorize customers into age groups.




labels = ['Yound Adult','Adult','Middle-Age','Senior']
```

```python
# This line creates a new column called 'age_group' in the DataFrame.
# It uses `pd.qcut` to divide the 'age' column into 4 equal-sized groups (quantiles).
# The `labels` list (defined previously) is then used to name these age groups:
# 'Young Adult', 'Adult', 'Middle-Age', and 'Senior'.


df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
```

```python
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
